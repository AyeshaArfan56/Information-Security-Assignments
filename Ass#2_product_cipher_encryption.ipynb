```python
plain_text = "ARRANGE A TRIP AFTER THE EXAMS"
alphabets = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"


def plain_to_cieser(plain_text , alphabets):
    cieser_text = ""
    
    
    plain_text = list(plain_text.upper())
    alphabets = list(alphabets.upper())
    
    
    
    for i in plain_text:
        
        if i.isalpha():
            
            index = (alphabets.index(i) + 3) % 26
            cieser_text = cieser_text + alphabets[index]
            
    return cieser_text



def cieser_to_vigenere(cieser_text , alphabets):
    key = "ayesha"
    key = list(key.upper())
    alphabets = list(alphabets.upper())
    cieser_text = list(cieser_text.upper())
    viegenere_text = ""
    
    
    for i in range(len(cieser_text) - len(key)):
        key.append(key[i % len(key)])
    
    for i in range(len(cieser_text)):
        
        index = (alphabets.index(cieser_text[i]) + alphabets.index(key[i])) % 26
        viegenere_text = viegenere_text + alphabets[index]
        
    return viegenere_text
        

def viegenere_to_auto_key(viegenere_text , alphabets):
    key = "bibi"
    key = list(key.upper())
    
    viegenere_text = list(viegenere_text.upper())
    
    auto_key = ""
    
    for i in range(len(viegenere_text) - len(key)):
        key.append(viegenere_text[i % len(viegenere_text)])
        
        
    for i in range(len(viegenere_text)):
        
        index = (alphabets.index(viegenere_text[i])+ alphabets.index(key[i])) % 26
        auto_key = auto_key + alphabets[index]
        
    return auto_key
        
    
    

        
        
cieser_text = plain_to_cieser(plain_text , alphabets)
print("Plain Text :",plain_text)
print("Cieser Text :",cieser_text)

viegenere_text = cieser_to_vigenere(cieser_text , alphabets)
print("Viegenere Text :",viegenere_text)


auto_key = viegenere_to_auto_key(viegenere_text , alphabets)
print("Auto Key :",auto_key)
        
```
